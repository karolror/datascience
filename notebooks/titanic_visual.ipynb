```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('titanic.csv')
```

```python
df.head(10)
```

```python
df.isna().sum()
```

```python
df.info()
```

### Age distribution plot for all passengers

```python
sns.displot(df, x="Age", kind="kde", fill=True)
plt.title('Age distribution chart')
```

- The graph deviates from the normal distribution
- The field tendency is right-skewed
- Most people are between 20 and 30 years old
- The fewest are the elderly from 70-80

### Histograms for subsets

```python
men = df.loc[df.Sex == 'male']["Age"]
women = df.loc[df.Sex == 'female']["Age"]
```

```python
plt.figure(figsize=(7, 7))
plt.subplot(2, 1, 1)
plt.hist(men)
plt.xlabel('Age')
plt.title('The age of the men')
plt.subplot(2, 1, 2)
plt.hist(women)
plt.xlabel('Age')
plt.title('The age of the women')
plt.subplots_adjust(hspace=0.35)
plt.show()
```

### Chart of dependence of ticket price on age

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=df, x='Age', y='Fare')
plt.yticks(np.arange(0, 300, step=20))
plt.xticks(np.arange(0, 80, step=5))
plt.show()
```

### Subplot

```python
K1= df[df['Pclass']==1]
K1_fare=K1['Fare'] #upper left histogram

K2= df[df['Pclass']==2]
K2_fare=K2['Fare'] # bottom left histogram

K3= df[df['Pclass']==3] #to the scatterplot on the right
```

```python
plt.figure(figsize=(10,10))

plt.subplot(2, 2, 1)
plt.title('Graph of tytanic data')
plt.hist(K2_fare, color='yellow')
plt.xlabel('II class fare')

plt.subplot(2, 2, 3)
plt.hist(K1_fare, color='red')
plt.xlabel('I class fare')

plt.subplot(1, 2, 2)
x=K3['Age']
y=K3['Fare']
plt.scatter(x, y, alpha=0.2)
plt.xlabel('Age of III class passengers')
plt.ylabel('Fare')
plt.grid()



plt.show()
```

### Boxplot for the age of disaster survivors and thosewho did not survive

```python
survived = df.loc[df['Survived'] == 1]
s_age = survived['Age']
nosurvived = df.loc[df['Survived'] == 0]
nos_age = nosurvived['Age']
```

```python
plt.figure(figsize=(10, 5))
plt.boxplot(s_age, positions = [1])
plt.boxplot(nos_age,positions = [2])
plt.show()
```

- medium in both cases in neighborhoods of 26-28
- more outliers in the second group
